```python
import generate_model
import heatx_network
from pyomo.environ import ConcreteModel, SolverFactory
```

```python
[n_CH4f, n_H2Of, n_O2f, n_N2f, Uf] = generate_model.feed_exhaust(Uf=0.8, Ua=0.2, MS=2, n_CH4f=10, n_H2ex=2)
```

```python
m = ConcreteModel()
```

```python
m = generate_model.concrete_model_base(m)
```

```python
m = generate_model.declare_material_streams(m, n_CH4f, n_H2Of, n_N2f, n_O2f, Uf)
```

```python
m = generate_model.set_constraint(m, air_in=700, fuel_in=500, ex_out=800)
```

```python
m.fs.Mix1.initialize()
m.fs.Mix2.initialize()
m.fs.Mix3.initialize()
m.fs.HX1.initialize()
m.fs.HX2a.initialize()
m.fs.HX2b.initialize()
m.fs.Split1.initialize()
```

```python
solver = SolverFactory('ipopt')
results = solver.solve(m, tee=True)
```

```python
generate_model.print_out(m, n_CH4f, LHV=50000*16.04, EE=0.9)
```

```python
h = ConcreteModel()
```

```python
h = heatx_network.declare_heatx_unit(h)
```

```python
h = heatx_network.define_heatx_streams(m, h)
```

```python
h = heatx_network.heatx_constraints(m, h, n_H2Of, enthalpy_vap=43988, air_in=700)
```

```python
solver = SolverFactory('ipopt')
results = solver.solve(h, tee=True)
```

```python
heatx_network.print_out_heatx(h)
```
